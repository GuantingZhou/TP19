## Scrapping a table from a website:


```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

## Data import for activity

```python
# Creating a function to import data
def data_import_clean(url, p=13, q=500, r=0, s=100):
    # Importing data
    headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',}
    response_activity = requests.get(url, headers=headers)
    soup_activity = BeautifulSoup(response_activity.text, "lxml")
    # finding column headers from imported data
    header_activity = []
    for i in soup_activity.find_all('b'):
        title = i.text
        header_activity.append(title)
        headers_activity = [x.replace('\n', '') for x in header_activity]
        headers_activity = [x.replace(' ', '') for x in  headers_activity]
        headers_activity = headers_activity[r:s]
        headers_activity
    data = pd.DataFrame(columns = headers_activity) # placing the column headers in a dataframe
    #populating the dataframe for each column
    for j in soup_activity.find_all('tr')[p:q]:
        row_data = j.find_all('td')
        row = [i.text for i in row_data]
        length = len(data)
        data.loc[length] = row
    # data cleaning
    data= data.apply(lambda s: s.str.strip())
    data= data.apply(lambda s: s.str.replace('\n',''))
    # Confirming all numeric columns are numeric
    data['130lb'] = pd.to_numeric(data['130lb'],errors='coerce')
    data['155lb'] = pd.to_numeric(data['155lb'],errors='coerce')
    data['180lb'] = pd.to_numeric(data['180lb'],errors='coerce')
    data['205lb'] = pd.to_numeric(data['205lb'],errors='coerce')
    # creating a new column named 'Average_Calorie' with the average value of all numeric columns for each row
    data['activity_calorie'] = data[['130lb','155lb','180lb','205lb']].mean(axis=1)
    # Showing the created dataframe
    return data
```

## Concatenating all dataframes to a single one named 'activity_data'

```python
activity_data = pd.concat([walking_data,swimming_data, cycling_data, dancing_excercise, hiking_data],ignore_index=True)
```

```python
# creating a new column named 'activity_id' with table index value
activity_data['activity_id'] = activity_data.index
activity_data
```

## Creating activity table

```python
# Selecting only required columns
activity = activity_data[['activity_id','activity', 'activity_calorie','activity_type_id']]
```

```python
activity
```

## Data cleaning of activity dataframe

```python
activity['activity']= activity['activity'].replace(r'\s+', ' ', regex=True)
```

```python
activity['activity']= activity['activity'].apply(lambda x: x.replace(',',''))
activity['activity']
```

```python
activity
```

```python
activity_one_times = activity.assign(activity = activity.activity + ' ; one day weekly')
```

```python
activity_one_times
```

```python
activity_two_times = activity.assign(activity = activity.activity + ' ; two days weekly')
activity_two_times['activity_calorie'] = activity['activity_calorie']*2
activity_two_times['activity_id'] = activity['activity_id']+79
```

```python
activity_two_times
```

```python
activity_three_times = activity.assign(activity = activity.activity + ' ; three days weekly')
activity_three_times['activity_calorie'] = activity['activity_calorie']*3
activity_three_times['activity_id'] = activity_two_times['activity_id']+79
activity_three_times
```

```python
activity_four_times = activity.assign(activity = activity.activity + ' ; four days weekly')
activity_four_times['activity_calorie'] = activity['activity_calorie']*4
activity_four_times['activity_id'] = activity_three_times['activity_id']+79
activity_four_times
```

```python
activity_five_times = activity.assign(activity = activity.activity + ' ; five days weekly')
activity_five_times['activity_calorie'] = activity['activity_calorie']*5
activity_five_times['activity_id'] = activity_four_times['activity_id']+79
activity_five_times
```

```python
activity_six_times = activity.assign(activity = activity.activity + ' ; six days weekly')
activity_six_times['activity_calorie'] = activity['activity_calorie']*6
activity_six_times['activity_id'] = activity_five_times['activity_id']+79
activity_six_times
```

```python
activity_7_times = activity.assign(activity = activity.activity + ' ; every day in a week')
activity_7_times['activity_calorie'] = activity['activity_calorie']*7
activity_7_times['activity_id'] = activity_six_times['activity_id']+79
activity_7_times
```

```python
activity_weekly = pd.concat([activity_one_times, activity_two_times, activity_three_times, activity_four_times, activity_five_times, activity_six_times, activity_7_times], ignore_index=True)
```
